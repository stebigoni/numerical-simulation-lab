```python
import numpy as np
import matplotlib.pyplot as plt

#Expected values (Black-Scholes)
call_exp = 14.975790778311286
put_exp = 5.4595325819072364



output = open("call_dir.dat","r")
M = int(output.readline())
output.close()

data = np.loadtxt(fname = "call_dir.dat",skiprows=1)
call = data[:,0]
call_err = data[:,1]

data = np.loadtxt(fname = "put_dir.dat",skiprows=1)
put = data[:,0]
put_err = data[:,1]

x = np.arange(1,np.size(call)+1)

plt.figure(figsize=(14,10))
plt.suptitle("European options prices")

plt.subplot(2,2,1)
plt.title("Call (direct sampling)")
plt.errorbar(x,call,call_err)
plt.axhline(y=call_exp, color='r', linestyle='-')


plt.subplot(2,2,2)
plt.title("Put (direct sampling)")
plt.errorbar(x,put,put_err)
plt.axhline(y=put_exp, color='r', linestyle='-')


data = np.loadtxt(fname = "call_disc.dat",skiprows=1)
call = data[:,0]
call_err = data[:,1]

data = np.loadtxt(fname = "put_disc.dat",skiprows=1)
put = data[:,0]
put_err = data[:,1]


plt.subplot(2,2,3)
plt.title("Call (discrete sampling)")
plt.errorbar(x,call,call_err)
plt.axhline(y=call_exp, color='r', linestyle='-')


plt.subplot(2,2,4)
plt.title("Put (discrete sampling)")
plt.errorbar(x,put,put_err)
plt.axhline(y=put_exp, color='r', linestyle='-')

plt.show()
```
