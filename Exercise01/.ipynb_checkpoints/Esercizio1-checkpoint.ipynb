Estimation of the mean value of the pseudo-random distribution and its error:

```python
import matplotlib.pyplot as plt
import numpy as np

output = open("output.dat","r")
M = int(output.readline())
output.close()

data = np.loadtxt(fname = "output.dat",skiprows=1)

r = data[:,0]
err = data[:,1]

N = np.size(r)

x = np.linspace(1,M,N)

plt.errorbar(x,r-0.5,yerr=err,color='orangered')
plt.title("Estimation of $<r>$")
plt.xlabel("# of throws")
plt.ylabel("$<r>$-1/2")
plt.grid(True)
plt.show()
```
